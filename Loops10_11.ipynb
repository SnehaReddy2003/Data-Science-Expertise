```python
i=1
while i<=5:
    print("DS", end=" ")
    j=1
    while j<=4:
        print("Technology", end= " ")
        j=j+1
    i=i+1
    print()
    
## the above cosde is more to make it optimisied so use "END"
## end=" " ------> this gives the same line instaed of next line
```

BREAK , CONTINUE, PASS

Break -----> it breaks the loop

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(1,11):
    if i==5:
        break
    print(i)
```

Continue: it skip the position and continue

```python
for i in range(1,11):
    if i==5:
        continue
    print(i)
```
